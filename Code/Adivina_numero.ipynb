```python
import random
import openpyxl
from openpyxl import Workbook
from openpyxl.worksheet.worksheet import Worksheet
import getpass
```

```python
def mostrar_menu_principal (nombre:str)->int:
    """Esta función pinta un mensaje personalizado de bienvenida, muestra las opciones del juego, recoge la opción escogida, 
    en caso de que no se ponga una de las opciones descritas, vuelve a pedirlo hata que sea correcta. Finalmente devuelve un 
    entero, el cual es la opción
    """
    print(f"     {nombre} BIENVENID@ AL JUEGO     \n"
          "1. Partida modo solitario\n2. Partida 2 Jugadores\n"
          "3. Estadística\n4. Salir\n")
    opcion:int = int(input("Escoja una opción:  "))
    while (opcion<1 or opcion>4):
        opcion = int(input("Escoja una opción válida:  "))

    return opcion
```

```python
def mostrar_submenu() -> int:
    print("1. Fácil (20 intentos)\n2. Medio (12 intentos)\n3. Difícil (5 intentos)")
    print()
    nivel:int = int(input("Seleccione un nivel de dificultad:   "))

    while nivel < 1 or nivel > 3:
        nivel = int(input("Seleccione un nivel de dificultad válido :   "))

    if nivel == 1:
        print("ES HORA DE EMPEZAR A JUGAR. Tienes 20 intentos. ")
    if nivel == 2:
        print("ES HORA DE EMPEZAR A JUGAR. Tienes 12 intentos. ")
    else:
        print("ES HORA DE EMPEZAR A JUGAR. Tienes 5 intentos. ")

    print()
    return nivel
```

```python
def calcular_numero_solitario() -> int:
    print("Partida solitario")
    numero: int = random.randint(1, 1000)
    return numero
```

```python
def calcular_2_jugadores(nombre:str) -> int:
    print("Partida de 2 jugadores.\n")
    num_secreto:int = int(getpass.getpass(f"Escriba un número para que {nombre} lo adivine: "))
    print()
    while not (num_secreto >= 1 and num_secreto <= 1000):
        print("El número debe estrar entre 1 y 1000")
        print()
        num_secreto:int = int(getpass.getpass(f"Escriba un número para que {nombre} lo adivine: "))

    print()
    return num_secreto
```

```python
def empezar_a_jugar(intentos: int, num_secreto: int, nombre_jugador: str) -> bool:
    for i in range(0, intentos):

        n_adivinando = int(input(f"Intento {i + 1}.\n   {nombre_jugador} ¿Cuál crees que es el número secreto?:\n"))
        if n_adivinando < 1 or n_adivinando > 1000:
            print("Sólo números del 1 al 1000")
            continue
        if (num_secreto == n_adivinando):
            print(f"ENHORABUENA {nombre_jugador},  ADIVINASTE.\n Has ganado en el intento {i}")
            print("\n----------------------------\n")
            return True
        elif (num_secreto < n_adivinando):
            print("El número secreto es menor que el número introducido.")
        else:
            print("El número secreto es mayor que el número introducido. ")

        if (i == intentos - 1):
            print(f"{nombre_jugador}! HAS PERDIDO. Se acabaron los intentos\nEl número secreto era {num_secreto}")
            print("\n----------------------------\n")
            return False
            
```

```python
documento_excel = openpyxl.load_workbook("./datos_juego.xlsx")
mi_hoja:Worksheet = documento_excel["Hoja1"]
diccionario = {1: 20, 2: 12, 3: 5}
ejecucion: bool = True
contador = 1

while (ejecucion):

    print("        -------ADIVINA ADIVINADOR-------             \n")
    nombre: str = input("Hola, dime tu nombre: ").upper()
    print()
    eleccion: int = mostrar_menu_principal(nombre)
    print()
    if eleccion == 1:
        num_a_adivinar: int = calcular_numero_solitario()
        nivel: int = mostrar_submenu()
        print("La máquina ha generado un número secreto.")
        print()
        ganar_o_perder: bool = empezar_a_jugar(diccionario[nivel], num_a_adivinar, nombre)
        mi_hoja.append([nombre,ganar_o_perder])
        documento_excel.save("./datos_juego.xlsx")
        contador = contador + 1
        
    elif eleccion == 2:
        nivel: int = mostrar_submenu()
        num_a_adivinar: int = calcular_2_jugadores(nombre)
        print(f"{nombre} Empieza tu turno.")
        ganar_o_perder = empezar_a_jugar(diccionario[nivel], num_a_adivinar, nombre)
        mi_hoja.append([nombre,ganar_o_perder])
        documento_excel.save("./datos_juego.xlsx")

        contador = contador + 1
        

    elif eleccion == 3:
        celdas = mi_hoja["A1":"B"+str(contador)]
        for row in celdas:
            for i in row:
                print(f"{i.value}  ", end="")
            print()


    elif eleccion == 4:

        ejecucion = False
```
